```python
%matplotlib inline
```

```python
import pandas as pd 
unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE']) #convert the DATE column into a series of datetime values 
print(unrate.head(12))
```

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

```python
first_twelve = unrate[0:12]
plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
plt.show()
```

```python
plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
plt.xticks(rotation=90) #rotate x-axis by 90 degrees
plt.show()
```

```python
plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
plt.xticks(rotation=90)
plt.xlabel('Month') #change x-axis label 
plt.ylabel('Unemployment Rate') #change y-axis label 
plt.title('Monthly Unemployment Trends, 1948') #set plot title
plt.show()
```
